```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\\Data_Analysis_Projects\\Company_Sales_Analysis\\raw_data\\Details.csv")
df.head()
```

Some required functions -

```python
import random

def generate_random_colors(n):
    colors = []
    for _ in range(n):
        color = "#{:06x}".format(random.randint(0, 0xFFFFFF))
        while color in colors:
            color = "#{:06x}".format(random.randint(0, 0xFFFFFF))
        colors.append(color)
    return colors

# Example usage:
n = 5
random_colors = generate_random_colors(n)
print(random_colors)
```

Task_1: On how many order the company have made profit? (Show using graph)

```python
#collecting data
count_profit = df.loc[df["Profit (per Q.)"] > 0].shape[0]
count_no_profit_loss = df.loc[df["Profit (per Q.)"] == 0].shape[0]
count_loss = df.loc[df["Profit (per Q.)"] < 0].shape[0]
```

```python
#plotting the graph
# plt.figure(dpi = 300)
plt.bar(["With Profit", "With Loss", "With 0 profit/loss"], [count_profit, count_loss, count_no_profit_loss], color = generate_random_colors(3))

plt.title("Graph showing orders with profit or loss os 0 profit/loss", fontdict = {"fontsize" : 14, "fontname" : "Comic Sans MS"})
plt.xlabel("Category", fontdict = {"fontsize" : 14, "fontname" : "Comic Sans MS"})
plt.ylabel("No. of Orders", fontdict = {"fontsize" : 14, "fontname" : "Comic Sans MS"})

plt.show()
```

Task_2: From our company, which category of products in at high demand?

```python
#collecting the data
category_dict = df["Category"].value_counts().to_dict()
print(category_dict)
```

```python
#plotting collected data on the graph
# plt.figure(dpi = 300)
plt.bar(category_dict.keys(), category_dict.values(), color = generate_random_colors(len(category_dict)))

plt.title("Graph showing no. of orders from different categories", fontdict = {"fontsize" : 15, "fontname" : "Comic Sans MS"})
plt.xlabel("Category", fontdict = {"fontsize" : 14, "fontname" : "Comic Sans MS"})
plt.ylabel("No. of Orders", fontdict = {"fontsize" : 14, "fontname" : "Comic Sans MS"})

plt.show()
```

Plot a graph from the amount or prices of sales of different categories.

```python
#colleting the data
category_sales_amount_dict = df.groupby("Category").agg({"Final Amount" : "sum"}).to_dict()
print(category_sales_amount_dict)
```

```python
#plotting the graph
# plt.figure(dpi = 300)
plt.bar(category_sales_amount_dict["Final Amount"].keys(), category_sales_amount_dict["Final Amount"].values(), color = generate_random_colors(len(category_sales_amount_dict["Final Amount"])))

plt.title("Graph showing total sales amount from different categories", fontdict = {"fontsize" : 15, "fontname" : "Comic Sans MS"})
plt.xlabel("Category", fontdict = {"fontsize" : 14, "fontname" : "Comic Sans MS"})
plt.ylabel("No. of Orders", fontdict = {"fontsize" : 14, "fontname" : "Comic Sans MS"})
plt.show()
```
