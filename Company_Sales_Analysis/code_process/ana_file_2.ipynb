```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
orders = pd.read_csv("D:\\Data_Analysis_Projects\\Company_Sales_Analysis\\raw_data\\Orders.csv")
details = pd.read_csv("D:\\Data_Analysis_Projects\\Company_Sales_Analysis\\raw_data\\Details.csv")
```

```python
#merging both files
df = pd.merge(left = orders, right = details, how = "inner", on = "Order ID")
df.head()
```

Some required functions - 

```python
import random

def generate_random_colors(n):
    colors = []
    for _ in range(n):
        color = "#{:06x}".format(random.randint(0, 0xFFFFFF))
        while color in colors:
            color = "#{:06x}".format(random.randint(0, 0xFFFFFF))
        colors.append(color)
    return colors

# Example usage:
n = 5
random_colors = generate_random_colors(n)
print(random_colors)
```

Task_3: Show count of orders from each state.

```python
#collecting data
state_dict = df["State"].value_counts().to_dict()
print(state_dict)
```

```python
#plotting the graph
plt.figure(figsize = (36, 18)) #dpi = 300
plt.bar(state_dict.keys(), state_dict.values(), color = generate_random_colors(len(state_dict)))

plt.title("Orders from every state", fontdict = {"fontsize" : 28, "fontname" : "Comic Sans MS"})
plt.xlabel("State", fontdict = {"fontsize" : 24, "fontname" : "Comic Sans MS"})
plt.ylabel("No. of Orders", fontdict = {"fontsize" : 24, "fontname" : "Comic Sans MS"})

plt.show()
```
